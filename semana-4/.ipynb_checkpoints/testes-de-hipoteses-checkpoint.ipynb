### Entregável obrigatório ###

Formular 1 hipótese de negócio (ex.: “taxa de resposta da campanha A = campanha B”) e descrever como testaria isso (sem precisar fazer o teste exato).

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
from statsmodels.stats.proportion import proportions_ztest
```

```python
df = pd.read_csv("marketing_campaign.csv", sep=";")
```

```python
df.info()
```

```python
# Valores únicos de escolaridade

unicos = df['Education'].unique()
unicos
```

```python
# Taxa de resposta de quem não tem ensino superior 

sem_sup = df[df['Education'].isin(['Basic','2n Cycle'])]
#print(sem_sup.head())
```

```python
# Obtenção da taxa média

tx_sem_sup = sem_sup['Response'].mean()
print("Taxa de resposta (SEM Ensino Superior):", round((tx_sem_sup)*100,1),"%")
```

```python
# Taxa de resposta de quem tem ensino superior 

com_sup = df[df["Education"].isin(['Graduation', 'PhD', 'Master'])]
#print(com_sup.head())
```

```python
# Obtenção da taxa média

tx_com_sup = com_sup["Response"].mean()
print("Taxa de resposta (COM Ensino Superior):", round((tx_com_sup)*100,1),"%")
```

```python
# Grupo com ensino superior

resposta_com = com_sup['Response'].sum()
total_com = com_sup.shape[0]

# Grupo sem ensino superior

resposta_sem = sem_sup['Response'].sum()
total_sem = sem_sup.shape[0]

# Arrays
respostas = np.array([resposta_com, resposta_sem])
totais = np.array([total_com, total_sem])

# Teste Z
stat, p = proportions_ztest(count=respostas, nobs=totais)

print("Estatística Z:", stat)
print("p-valor:", p)

# Interpretação automática

if p < 0.05:
    print("Rejeitamos H0: há evidências de diferença entre grupos.")
else:
    print("Não rejeitamos H0: não há evidência suficiente de diferença.")
```
