# Normalização e Padronização # 

O escalonamento de atributos **(feature scaling)** é uma das transformações mais críticas no pré-processamento de dados, pois a maioria dos algoritmos de Machine Learning apresenta um desempenho insatisfatório quando os atributos numéricos de entrada possuem escalas muito distintas.

Existem duas abordagens principais para uniformizar as escalas: a **normalização** e a **padronização.**


### 1. Normalização (Min-Max Scaling) ###

A normalização, frequentemente referida como *Min-Max scaling*, redimensiona os valores de modo que eles passem a variar dentro de um intervalo fixo, geralmente entre 0 e 1.

- **Como funciona**: O processo subtrai o valor mínimo do atributo e divide o resultado pela amplitude total (valor máximo menos o valor mínimo). No Scikit-Learn, essa técnica é implementada pelo transformador *MinMaxScaler*, que permite inclusive alterar o intervalo desejado (ex: -1 a 1) por meio do hiperparâmetro *feature_range*.


- **Senbilidade a Outliers**: A normalização é **extremamente sensível a valors atípicos** (outliers). Por exemplo, se houver um erro em um conjunto de dados de renda onde um valor de 100.000 é registrado em vez de 10.000, a normalização irá "esmagar" todos os outros valores legítimos (que variam entre 0 e 15.000) para um intervalo minúsculo (como 0 a 0,15), dificultando a detecção de padrões pelo modelo.


- **Quando usar**: É indicada quando o algoritmo de destino exige especificamente um intervalo fixo, como em **Redes Neurais**, que frequentemente esperam entradas entre 0 e 1.


### 2. Padronização (Stardardization) ###

A padronização transforma os dados de forma que eles passem a ter **média zero** e **variância unitária** (desvio padrão igual a 1).

- **Como funciona**: Primeiro, subtrai-se a média das observações (centrando os dados em zero) e, em seguida, divide-se pelo desvio padrão. Estatisticamente, isso equivale ao cálculo da **variável padronizada** Z (Z=(X−μ)/σ). No Scikit-Learn, utiliza-se o transformador *StandardScaler*.


- **Vantagem sobre Outliers**: Diferente da normalização, a padronização não restringe os valores a um intervalo específico, o que a torna **muito menos afetada por outliers.** No caso de um erro de preenchimento, a padronização manteria a maior parte dos dados em uma escala útil, tratando o valor discrepante apenas como um ponto com desvio padrão muito elevado.


- **Uso Estatístico**: É fundamental para o cálculo do **coeficiente de correlação**, que é definido como a média dos produtos dos valores padronizados das variáveis.


### Modelos Adequados e Necessidade de Escalonamento ###

O escalonamento não é opcional para muitos modelos; ele é um requisito para o funcionamento correto e eficiente dos seguintes algoritmos:

- **Modelos baseados em Gradiente Descendente**: Algoritmos como **Regressão Linear**, **Regressão Logística** e **Redes Neurais** convergem muito mais rápido para a solução ideal se os atributos estiverem na mesma escala. Sem isso, o algoritmo pode levar muito mais tempo para atingir o mínimo global da função de custo.


- **Algoritmos Baseados em Distância**: Modelos como k-Nearest Neighbors (k-NN) e K-Means calculam distâncias (como a Euclidiana) entre pontos. Se um atributo tiver uma magnitude muito maior que os outros (ex: renda em milhares vs. idade em dezenas), ele dominará completamente o cálculo da distância, tornando os outros atributos irrelevantes.


- **Support Vector Machines (SVMs)**: São altamente sensíveis às escalas. Sem o escalonamento, o SVM pode negligenciar atributos com pequenas magnitudes e focar apenas nos de maior valor, o que compromete a criação da "margem larga" de separação.


- **Análise de Componentes Principais (PCA)**: Como o PCA busca identificar as direções de maior variância, atributos com escalas maiores podem parecer ter mais variância do que realmente têm, distorcendo os componentes principais.


- **Modelos com Regularização (Ridge e Lasso)**: Como esses modelos adicionam uma penalidade baseada na magnitude dos pesos dos coeficientes, atributos em escalas maiores seriam penalizados injustamente, levando a um modelo enviesado.




## Regra de Ouro do Professor: ## 
Você deve ajustar (fit) os escalonadores **apenas nos dados de treinamento.** Jamais utilize o conjunto de teste ou o dataset completo para o ajuste, pois isso causaria o vazamento de informação (data leakage), fornecendo ao modelo conhecimentos sobre a distribuição global dos dados que ele não deveria ter durante o treino. Somente após o ajuste no treino, utilize o escalonador para transformar os dados de teste e novos dados.

# Features Importances # 

A **feature importance** (ou importância de atributos) é uma técnica fundamental na Ciência de Dados que permite avaliar o peso ou a relevância de cada variável de entrada para o resultado final das previsões de um modelo.


### 1. O que é "Feature Importance"? ###

Em termos simples, ela nos diz **quais variáveis "chamam mais a atenção" do modelo** na hora de decidir. Tecnicamente, em modelos baseados em árvores, a importância de um atributo é medida pelo quanto os nós que utilizam essa variável conseguem **reduzir a impureza** (como o índice de Gini) em média ao longo de todas as árvores do modelo. O resultado é geralmente apresentado como uma pontuação (score) para cada variável, onde a soma de todas as importâncias é igual a 1 (ou 100%). **Por exemplo**, ao prever preços de casas, o modelo pode indicar que a "renda média" tem uma importância de 0,36 (36%), sendo o fator mais decisivo.


### 2. Quando utilizar? ###
A importância dos atributos deve ser utilizada em três momentos principais:

- **Seleção de Variáveis (Feature Selection):** Ajuda a identificar e remover variáveis irrelevantes ou redundantes que não contribuem para o modelo, o que simplifica o pipeline e pode melhorar a performance.


- **Entendimento de Negócio (Insights):** Permite ao cientista de dados explicar de forma rápida para as partes interessadas o que realmente importa nos dados para aquele problema específico. 


- **Limpeza e Depuração:** Ao inspecionar as importâncias, você pode perceber que o modelo está focando em variáveis que podem estar causando vazamento de dados *(data leakage)* ou que possuem muitos valores atípicos *(outliers)* distorcendo o padrão.


### 3. Quais modelos possuem "Feature Importance"? ###

Embora o conceito de relevância exista em quase toda a modelagem estatística (como nos coeficientes β da regressão linear
), a métrica específica *feature_importances_* é característica de modelos não-lineares, especialmente os **Ensembles:**


- **Random Forests:** É o modelo mais citado para essa técnica, onde a importância é calculada automaticamente após o treinamento.


- **Decision Trees (Árvores de Decisão):** Como são a base das florestas, elas também fornecem essa medida individualmente.


- **Extra-Trees:** Seguem a mesma lógica das florestas aleatórias para medir a redução de impureza.


- **Modelos de Boosting (AdaBoost, Gradient Boosting, XGBoost):** Também utilizam essa métrica para avaliar quais atributos melhor corrigiram os erros das árvores anteriores no processo sequencial.


## Dica de Professora: ## 
Na estatística clássica, para saber se uma variável é importante em uma regressão, costumamos olhar para o **coeficiente de explicação (R²)** ou realizar testes de correlação para ver o quanto uma variável explica a variação da outra. No entanto, no Machine Learning moderno, os modelos de florestas são os favoritos dos analistas para gerar esses rankings de importância de forma automatizada e robusta.

# Divisão TimeSeriesSplit #

A **Divisão TimeSeriesSplit (ou Divisão de Séries Temporais)** é uma técnica de validação cruzada específica para lidar com dados que possuem uma ordem cronológica, onde as observações passadas influenciam as futuras. De acordo com as fontes, os principais fundamentos dessa técnica são:


### 1. A Natureza dos Dados Temporais ###

Diferente de conjuntos de dados independentes (como notas de alunos), em uma série temporal as observações são realizadas em instantes ordenados do tempo. Existe uma dependência temporal, o que significa que o valor de hoje é frequentemente relacionado aos valores de ontem. Exemplos típicos citados nas fontes incluem preços de ações, condições climáticas e índices de poluição.


### 2. O Problema do Embaralhamento (Shuffling) ###

Em modelos tradicionais de Machine Learning, costuma-se embaralhar os dados para garantir que os conjuntos de treino e teste sejam estatisticamente semelhantes. No entanto, as fontes alertam que **embaralhar dados de séries temporais é uma má ideia.**

- Se você embaralhar o tempo, o modelo pode acabar usando informações do "futuro" para prever o "passado" durante o treinamento, o que resultaria em uma estimativa de erro irrealista.


### 3. Validação "Out-of-Time" ###

A técnica descrita como **Out-of-Time** é a base do funcionamento do *TimeSeriesSplit*.

- **Funcionamento**: O conjunto de dados é dividido em períodos de tempo distintos. O modelo é treinado nos dados de um período anterior (passado) e testado nos dados de um período subsequente (futuro).

- **Simulação de Realidade:** Esse método simula o cenário real de produção, onde o objetivo é garantir que o modelo consiga generalizar para dados que ainda não ocorreram.


### 4. Estrutura de Janela em Expansão ###

Embora o termo técnico específico apareça em bibliotecas como Scikit-Learn, as fontes descrevem o conceito estatístico de que o pesquisador não pode selecionar dados aleatórios em séries temporais, pois elas são "levantamentos observacionais" onde a série é uma realização de um processo estocástico. Na prática, a divisão ocorre criando sucessivos pontos de corte no tempo:

1. Treina no período 1 → Testa no período 2.
2. Treina nos períodos 1 e 2 → Testa no período 3.
3. E assim por diante, sempre respeitando a cadeia progressiva de eventos.

Em resumo, a Divisão TimeSeriesSplit é **essencial para evitar o vazamento de dados** *(data leakage)* do futuro para o passado, garantindo que a validação do modelo respeite a ordem lógica do tempo identificada na estatística clássica.


# Divisão Estratificada # 

A **divisão estratificada** (ou amostragem estratificada) é uma técnica em que a população é dividida em subgrupos homogêneos, chamados de estratos, para garantir que a amostra final seja uma representação fiel do todo. 

Aqui estão os pontos principais para entender como ela funciona e por que é usada:

### 1. O Objetivo: Representatividade ###

O principal motivo para usar a divisão estratificada é evitar o **viesamento de amostragem** *(sampling bias).*

- Em uma amostragem puramente aleatória, existe o risco de, por puro azar, obtermos um conjunto de dados que não reflete a realidade da população (por exemplo, ter homens demais em uma pesquisa sobre a população geral).

- A divisão estratificada garante que as proporções de subgrupos importantes sejam mantidas tanto no treino quanto no teste.


### 2. Exemplo Prático (População e Gênero) ### 

Os materiais trazem um exemplo claro: se a população dos EUA é composta por **51,3% de mulheres** e **48,7% de homens**, uma pesquisa bem conduzida de 1.000 pessoas deve tentar manter exatamente essa proporção: **513 mulheres e 487 homens na amostra.**


### 3. Aplicação em Ciência de Dados ###

Na prática do Machine Learning, essa **técnica é fundamental** ao lidar com **datasets pequenos** ou **com classes desbalanceadas.**

- **Divisão Treino/Teste:** Ao usar ferramentas como o *StratifiedShuffleSplit* do Scikit-Learn, o cientista de dados garante que a categoria de interesse (como faixas de renda) tenha a mesma distribuição percentual no conjunto de treinamento e no conjunto de teste.

- **Redução de Variância:** Um dos objetivos estatísticos da estratificação é **homogeneizar** a variância dentro de cada estrato em relação à variável de interesse, o que torna as estimativas mais precisas.

### 4. Estratos na Prática Estatística ###

Conforme Morettin e Bussab, os estratos podem ser determinados por variáveis qualitativas ou categorias. Por exemplo, em uma população de estudantes, pode-se criar estratos baseados na classe social (A, B ou C) e, em seguida, realizar uma Amostragem Aleatória Simples (AAS) dentro de cada um desses grupos.


## Resumo da Professora: ## 
Enquanto a amostragem aleatória simples "chuta" os dados para os conjuntos de treino e teste esperando que a sorte ajude, a **divisão estratificada** é um processo deliberado para **"espelhar" a realidade**, garantindo que o seu modelo aprenda com uma fatia justa e proporcional do mundo real.
