# Entender relações entre variáveis #

Quando analisamos uma base, precisamos identificar como as variáveis se relacionam:

**• Correlação:** mede se duas variáveis numéricas se movem juntas (ex.: renda e gasto em vinhos).

    
    Código:
        
        df[['Income','MntWines']].corr()
        
**•	Associação categórica:** verificar se categorias influenciam resultados (ex.: estado civil e taxa de resposta).

**•	Dependência temporal:** variáveis ligadas ao tempo (ex.: Recency e Response) podem indicar comportamento recente.

Em linguagem de negócio: entender relações ajuda a descobrir **drivers de comportamento.** Exemplo: “Clientes com maior renda tendem a gastar mais em vinhos” → isso orienta segmentação de campanhas.

# Binning (agrupar variáveis contínuas em faixas) #

**Binning** é transformar uma variável contínua em categorias.

Exemplo: transformar Income em faixas → “Baixa renda”, “Média renda”, “Alta renda”.

**Vantagens:**
- Simplifica a interpretação.
- Útil para modelos que lidam com categorias.

**Riscos:**

**•	Perda de informação:** ao agrupar, você perde nuances (ex.: 4999 reais e 5001 reais podem cair em faixas diferentes).

**•	Criação artificial de padrões:** pode parecer que há diferenças significativas só porque os grupos foram definidos arbitrariamente.

**• Overfitting:** se o binning for feito com base no target (ex.: criar faixas de renda que maximizam taxa de resposta), você está “ensinando” o modelo com a resposta já embutida.

# Leakage (vazamento de informação) #

**Leakage** acontece quando uma variável usada no modelo contém informação que não estaria disponível no momento da previsão ou está diretamente ligada ao target.

Exemplos:
- Usar  (se o cliente respondeu) como variável para prever… justamente a resposta.
- Usar  se ele já foi influenciado pela campanha que você quer prever.
- Usar  (dias desde última compra) para prever se o cliente vai comprar, quando essa variável já reflete o comportamento futuro.

Em linguagem de negócio: leakage é como “trapacear” na previsão. Você está usando pistas que só existem depois do evento, o que gera modelos com performance irreal e inúteis na prática.

# Estratégias simples para evitar problemas#

1. 	Explorar relações com correlação, gráficos e tabelas de frequência antes de modelar.
2. 	Evitar binning arbitrário: se precisar, use métodos estatísticos (quantis, equal-width bins) e sempre valide se faz sentido para o negócio.
3. 	Checar leakage: pergunte sempre “essa informação estaria disponível no momento da previsão?”
- Se não, não use.
- Exemplo: para prever resposta a uma campanha, não use variáveis que já refletem a resposta
