## Tipos de Variáveis ##

**1. Variáveis Categóricas (Qualitativas)**

Representam características, qualidade ou categorias que não podem ser medidas numericamente, apenas organizadas em grupos. 

**Nomimais:** Categorias sem ordem ou ranking específico. **Exemplo de Negócio:** Gênero, Tipo de Produto, Cidade, Forma de Pagamento (crédito, débito, pix), categoria do cliente (VIP, Regular).


**Ordinais:** Categorias com uma ordem lógica ou ranking. **Exemplo de Negócio:** Nível de Satisfação do Cliente (baixo, médio, alto), Escolaridade, Tamanho da Empresa (Pequena, Média, Grande), Avaliação de Desempenho.



**2. Variáveis Numéricas (Quantitativas)**

Representam quantidades mensuráveis e podem ser manipuladas matematicamente (média, soma). 

**Discretas:** Valores inteiros contáveis, geralmente resultantes de contagens. Não aceitam frações ou casas decimais.**Exemplo de Negócio:** Número de clientes por dia, quantidade de itens vendidos, número de funcionários, total de reclamações.


**Contínuas:** Valores que assumem qualquer valor em um intervalo contínuo, geralmente mensurados (balança, régua, relógio). **Exemplo de Negócio:** Valor da Venda (R$), Peso do Produto, Tempo de Atendimento, Margem de Lucro (%).

### Entregável avançado ###

Carregar o dataset em Python (pandas) e **gerar automaticamente o dicionário de variáveis** (tipo inferido, % missing, exemplos de valores).

```python
# Carregar dataset
df
```

```python
# Função para gerar dicionário de variáveis 

def dicionario_variaveis(df, exemplos=4):
    dicionario = [] # inicializa uma lista vazia que vai guardar as infos de cada coluna
    for coluna in df.columns:
        tipo = df[coluna].dtype # tipo de variável
        missing = df[coluna].isna().mean()*100 # porcentagem de valores ausentes
        
        exemplos_valores = df[coluna].dropna().unique()[:exemplos] # exemplos únicos
        
        dicionario.append({
            "Variável": coluna,
            "Tipo": str(tipo),
            "% Missing": round(missing, 2),
            "Exemplos": exemplos_valores
        })
    return pd.DataFrame(dicionario)
```

```python
# Gerar dicionário 

dic = dicionario_variaveis(df)
dic
```
